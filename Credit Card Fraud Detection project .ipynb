```python
from sklearn.preprocessing import StandardScaler,RobustScaler
sts=StandardScaler()
rbs=RobustScaler()

data["Amount"]=sts.fit_transform(data['Amount'].values.reshape(-1,1))
data["Time"]=sts.fit_transform(data['Time'].values.reshape(-1,1))


data['Amount']=rbs.fit_transform(data['Amount'].values.reshape(-1,1))
data['Time']=rbs.fit_transform(data['Time'].values.reshape(-1,1))
```

```python
Amount=data['Amount']
Time=data['Time']

data.drop(['Amount','Time'],axis=1,inplace=True)

data.insert(0,'Amount',Amount)
data.insert(1,'Time',Time)

data.head()
```

```python
data= data.sample(frac=1)

# amount of fraud classes 492 rows.
fraud_data = data.loc[data['Class'] == 1]
non_fraud_data = data.loc[data['Class'] == 0][:492]

normal_distributed_data = pd.concat([fraud_data, non_fraud_data])

# Shuffle dataframe rows
new_data = normal_distributed_data.sample(frac=1, random_state=42)

new_data.head()
```

```python
print('Distribution of the Classes in the subsample dataset')
print(new_data['Class'].value_counts()/len(new_data))

sns.countplot('Class', data=new_data, palette=colors)
plt.title('Equally Distributed Classes', fontsize=14)
plt.show()
```

```python
#Corelation 
fig,(ax1,ax2)=plt.subplots(2,1,figsize=(24,20))

#Entire Dataframe
corr=data.corr()
sns.heatmap(corr,cmap='coolwarm_r',annot_kws={'size=20'},ax=ax1)
ax1.set_title('Imbalanced Correlation Matrix',fontsize=15)

sub_sample_corr=new_data.corr()
sns.heatmap(sub_sample_corr,cmap='coolwarm_r',annot_kws={'size=20'},ax=ax2)
ax2.set_title('SubSample Correlation Matrix',fontsize=15)
```
